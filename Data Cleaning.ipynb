```python
import pandas as pd
import numpy as np
import re
```

```python
def getLineEq(points):
    y1 = points[0][1]
    y2 = points[1][1]
    x1 = points[0][0]
    x2 = points[1][0]
    slope = (y2-y1)/(x2-x1)
    intercept = y1-slope*x1 
    return (slope, intercept)
```

```python
def removeChar(s):
    s = str(s)
    return float(re.sub('\D', '',s))/10
```

```python
df = pd.read_csv(r"C:\Users\Victor\iCloudDrive\San Francisco State University\googleHacks\Full Dataset - Final.csv")
```

```python
df
```

```python
char_col = ['labor force participation rate', 'employment-population ratio', 'labor force', 'employment', 
            'unemployment','unemployment rate']
```

```python
for c in char_col:
    df[c] = df[c].transform(removeChar)
```

```python
for col in df.columns[df.isna().any()]:
    df[col].interpolate(method='linear', inplace=True)
```

```python
for col in df.columns[df.isna().any()]:

    coordinates = []
    for x,y in enumerate(df[col]):
        if not np.isnan(y):
            coordinates.append((x+1,y))
        if len(coordinates)==2:
            break
    coordinates
    f = getLineEq(coordinates)
    for x,item in enumerate(df[col].isnull()):
        if item:
            y = f[0]*(x+1)+f[1]
            df.loc[x, col] = y
df
```
